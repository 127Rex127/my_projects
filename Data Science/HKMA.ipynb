```python
import requests
```

```python
url = 'https://api.hkma.gov.hk/public/bank-svf-info/fraudulent-bank-scams?lang=en'
```

```python
data = requests.get(url).json()
```

```python
data
```

```python
data['result']['records']
```

```python
for i in data['result']['records']:
    print('Issue Date:',i['issue_date'])
    print('Bank Name:',i['alleged_name'])
    print('Scam:',i['scam_type'])
    print('URL:',i['pr_url'])
    print('Scam Website:',i['fraud_website_address'])
    print('='*140)
    
```
