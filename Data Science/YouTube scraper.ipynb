```python
import feedparser
```

```python
channel = 'UCDogdKl7t7NHzQ95aEwkdMw'
```

```python
rss_url = f"https://www.youtube.com/feeds/videos.xml?channel_id={channel}"
```

```python
rss_url
```

```python
feed = feedparser.parse(rss_url)
```

```python
feed
```

```python
feed['entries']
```

```python
feed['entries'][0]['title']
```

```python
len(feed['entries'])
```

```python
for i in feed['entries']:
    print('Title:',i['title'])
    print('Link:',i['link'])
    print('Description:',i['description'][0:100])
    print('Views:',(f'{int(i['media_statistics']['views']):,}'))
    print('='*125)
```
